## Modifing the publishind Date and Price col
---

```python
books['Publish Date'] = books['Publish Date'].str.split(',').str[-1].str.strip()
# This code splits the publishing date string by commas, takes the last element (which contains the year), and removes any leading or trailing whitespace.

books['Price'] = books['Price'].str.replace('Price Starting at $', '', regex=False)
# This code replaces the string "Price Starting at $" with an empty string in the price column.

books
```

## Merge

---



```python
b1 = pd.read_csv('/content/drive/MyDrive/datasets/goodreads_books.csv',on_bad_lines='warn')
b2 = pd.read_csv('/content/drive/MyDrive/datasets/BooksDataset.csv',on_bad_lines='warn')
b1.columns = b1.columns.str.capitalize()
b1
```

```python
b2.head()
```

```python
b1.merge(b2, how = 'inner' , on = 'Title')
```

```python
b1.merge(b2, how = 'right' , on = 'Title')
```

```python
b1.merge(b2, how = 'cross')
```

```python
b1.merge(b2, how = 'left' , on = 'Title')
```
